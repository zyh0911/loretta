```python
import torch

# 原始的 a 和 b
a = torch.randn(14, 12)  # 大小为 (14, 12)
b = torch.randn(12, 6)   # 大小为 (12, 6)

# Step 1: 矩阵乘法 (14, 12) 和 (12, 6)
original_result = torch.matmul(a, b)  # 结果是 (14, 6)

# Step 2: 重塑 a 和 b 为 (14, 3, 4) 和 (3, 4, 6)
a_reshaped = a.view(14, 3, 4).permute(1, 0, 2) 
# print(a)
# print(a_reshaped)
# print(a.view(3, 14, 4))

b_reshaped = b.view(3, 4, 6)   # (3, 4, 6)

# Step 3: 计算分块乘法
result = torch.matmul(a_reshaped, b_reshaped)  # 结果是 (3, 14, 6)

# Step 4: 将分块结果沿着第一个维度求和，得到最终结果
final_result = result.sum(dim=0)  # 求和得到 (14, 6)

# 输出比较
print(torch.allclose(original_result, final_result))  # 检查是否一致
```
